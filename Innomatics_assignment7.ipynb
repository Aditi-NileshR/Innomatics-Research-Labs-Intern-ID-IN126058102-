2: Employee ID Card System

```python
class employees:
  def __init__(self,name,ID,dept):
    self.employee_name = name
    self.employee_ID = ID
    self.employee_department = dept

employee1 = employees("Rahul","EMP102","IT")
print("Employee ID Card \n ")
# Corrected printing of employee details
print(f"Name: {employee1.employee_name}")
print(f"ID: {employee1.employee_ID}")
print(f"Department: {employee1.employee_department}")
```

3: Mobile Contact Record


```python
class phonebook:
  def __init__(self):
    self.name = str(input("Contact name:"))
    self.number = int(input("Phone number:"))
  def o_p(self):
    print(f"\n Contact saved \n Name: {self.name} \n Phone number: {self.number}")
p1 = phonebook()
p1.o_p()
```

4: Product Price Tag Generator

```python
class price_tag:
  def __init__(self):
    self.name = str(input("Product name:"))
    self.price = int(input("Price:"))
  def o_p(self):
    print(f"\n Product name: {self.name} \n Price: {self.price}")
p1 = price_tag()
p1.o_p()
```

5: Movie Rating Display System

```python
class movie_ratings:
  def __init__(self):
    self.name = str(input("Movie name:"))
    self.rating = float(input("Rating:"))
  def o_p(self):
    print(f"\n Movie name: {self.name} \n Rating: {self.rating}/5")
p1 = movie_ratings()
p1.o_p()
```
