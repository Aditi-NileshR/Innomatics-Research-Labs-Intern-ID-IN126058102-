```python
import re
from collections import Counter

# ==========================================
# Task 1:Conceptual Understanding (Mandatory – Written)
# ==========================================
print("""1. What is the difference between Love and love in NLP?
In Natural Language Processing, Love and love are treated as two entirely distinct tokens by default because computers read them based on their differing ASCII/Unicode character values. If text is not normalized (converted to lowercase), a machine learning model will unnecessarily learn them as two separate words, diluting the frequency patterns and meaning of the word.

2. What happens if stopwords are not removed?
If stopwords highly common words like is, the, and) are left in the text, they introduce significant "noise" into the dataset. Because they appear so frequently, they can overwhelm the model, increase the dimensionality of the data, and consume computational memory without adding any real semantic value or meaning to tasks like document classification.

3. Provide two real-world scenarios where removing stopwords can be harmful.
Sentiment Analysis: Removing negative stopwords like "not" can completely reverse the sentiment of a sentence. For example, "The food was not good" would become "food good."
Machine Translation / Text Generation: These applications rely heavily on grammar, syntax, and sentence structure to generate human-readable text. Removing stopwords strips away the grammatical glue, resulting in broken and incoherent outputs.

4. Explain the difference between stemming and lemmatization.
Stemming is a blunt, rule-based approach that simply chops off the prefixes or suffixes of a word to find its root (e.g., chopping "studying" to "study", but mistakenly chopping "better" to "bett").
Lemmatization is a more advanced, context-aware approach that uses a vocabulary dictionary and morphological analysis to accurately return a word to its proper base dictionary form, known as a lemma (e.g., safely converting "better" to "good").""")

# ==========================================
# Task 2: Build Advanced Preprocessing Function
# ==========================================
def preprocess_text(text):
    # Task 7: Error Handling - Handle empty strings or non-string inputs
    if not isinstance(text, str) or not text.strip():
        return [], ""

    # 6. Remove URLs and email-like patterns
    text = re.sub(r'http[s]?://\S+|www\.\S+|\S+@\S+', '', text)

    # 4. Convert text to lowercase
    text = text.lower()

    # 1. Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove punctuation and emojis (Replace non-alphanumeric chars with spaces)
    # This prepares the text for clean tokenization
    text = re.sub(r'[^\w\s]', ' ', text)

    # 3. Handle repeated characters
    # Replaces 3 or more consecutive identical characters with 2 characters.
    # We reduce to 2 instead of 1 to preserve real words like "good" or "too".
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Tokenize text (this naturally handles '2. Remove extra spaces' as well)
    raw_tokens = text.split()

    # 5. Remove very short tokens (length <= 2) EXCEPT "no" and "not"
    exceptions = {'no', 'not'}
    cleaned_tokens = [word for word in raw_tokens if len(word) > 2 or word in exceptions]

    # Reconstruct the cleaned sentence
    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence

# ==========================================
# Task 6: Build Full Pipeline
# ==========================================
def full_pipeline(text_list):
    all_tokens = []
    all_clean_sentences = []

    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        all_tokens.extend(tokens) # Flattening the list for global analysis
        all_clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": all_clean_sentences
    }

# ==========================================
# Task 3: Stress Testing
# ==========================================
# Original sample inputs + Task 7 Error Handling edge cases
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this",
    "",               # Error Handling: Empty string
    "😀🚀🔥",           # Error Handling: Only emojis
    "123 456 7890"    # Error Handling: Only numbers
]

print("="*50)
print("TASK 3: STRESS TESTING & TASK 7: ERROR HANDLING")
print("="*50)

# ==========================================
# Task 4: Token Analytics
# ==========================================
sentence_analytics = []

for text in sample_inputs:
    tokens, clean_text = preprocess_text(text)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(t) for t in tokens) / total_tokens if total_tokens > 0 else 0

    # Store analytics for later questions
    sentence_analytics.append({
        "original": text,
        "clean": clean_text,
        "total_tokens": total_tokens
    })

    print(f"Original Text    : {text}")
    print(f"Cleaned Tokens   : {tokens}")
    print(f"Cleaned Sentence : {clean_text}")
    print(f"-> Analytics     : Total: {total_tokens} | Unique: {unique_tokens} | Avg Len: {avg_length:.2f}")
    print("-" * 50)

# Analyzing Task 4 Questions based on processed data:
print("\nTask 4 Analysis Answers:")
# The sentence with the most noise is the one that lost the most characters/tokens compared to its original length.
print("1. Which sentence had the most noise? ")
print("   -> 'Win $$$ now!!! Limited offer!!!' and 'Worst service ever... 0/10' contained highly irregular symbols, numbers, and punctuation.")
# The sentence with the most tokens remaining
most_meaningful = max(sentence_analytics, key=lambda x: x['total_tokens'])
print("2. Which sentence retained the most meaningful tokens after cleaning?")
print(f"   -> '{most_meaningful['original']}' (Retained {most_meaningful['total_tokens']} tokens)")

# ==========================================
# Task 5: Frequency Analysis
# ==========================================
print("\n" + "="*50)
print("TASK 5 & 6: PIPELINE RUN AND FREQUENCY ANALYSIS")
print("="*50)

# Run full pipeline to gather all tokens globally
pipeline_results = full_pipeline(sample_inputs)
global_tokens = pipeline_results["tokens"]

word_counts = Counter(global_tokens)

# Top 10 most frequent words
top_10 = word_counts.most_common(10)

# Top 5 least frequent words
# most_common() returns descending order, so we slice the reverse for least common
least_5 = word_counts.most_common()[:-6:-1]

print("Top 10 most frequent words:")
for word, count in top_10:
    print(f"  - '{word}': {count} times")

print("\nTop 5 least frequent words:")
for word, count in least_5:
    print(f"  - '{word}': {count} times")
```
