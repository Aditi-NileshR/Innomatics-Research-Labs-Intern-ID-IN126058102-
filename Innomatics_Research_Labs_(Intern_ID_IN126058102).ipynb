```python
#1. User Login Check

username = str(input("Entered_user"))
password = int(input("Entered_password"))
if username == "admin" and password == 1234:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#2. Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Number of Students Passed:", pass_count)
print("Number of Students Failed:", fail_count)
```

```python
#3. Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    new = name.strip().lower()
    cleaned_names.append(new)

print("Cleaned and Standardized usernames:", cleaned_names)
```

```python
#4. Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}'")
    print("Length:", length)

    if length > 10:
        print("Status: Long message (More than 10 characters)")
    else:
        print("Status: Short message")
```

```python
#5. Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0

for l in logs:
    if l == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
