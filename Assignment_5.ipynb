3. Agriculture-rainfall adequacy checker

```python
THRESHOLD = 50
def rainfall_checker(rainfall):
    avg_rainfall = round(sum(rainfall) / len(rainfall))
    if avg_rainfall > THRESHOLD:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")
    print(f"Average Rainfall:{avg_rainfall}")

rainfall = [60, 70, 80, 70, 70, 30, 90]
rainfall_checker(rainfall)
```

4. Social Media-Duplicate Account Detection

```python
acc = {"saurabh123", "srishti", "rahul78", "srishti", "rahul78", "kiran_564", "Areeba_99"}
def duplicate_account(acc):
    if len(acc) != len(set(acc)):
      print("Duplicate Account found : Yes")
    else:
      print("Duplicate Account found : No")
duplicate_account(acc)
```

6. Agriculture - Premium crop price filter

```python
threshold = 2000
premium_crops = []
def filter(crop_price):
  for r in crop_price:
    if r >= threshold:
      premium_crops.append(r)
  print(f"Premium Crops:{premium_crops}")
crop_price = list(map(int, input("Enter crop price: ").split()))
filter(crop_price)
```

7. System Monitoring - Application Health

```python
def error_count(errors):
  if errors == 0:
    print("System Status: Healthy")
  elif errors <= 5:
    print("System Status: Minor Issues")
  else:
    print("System Status: Critical Issues")
errors = int(input("Error Count:"))
error_count(errors)
```

8. Banking - Daily transaction limit checker

10. Smart Electricity Bill Analyzer

```python
def Bill(units):
  if units <= 100:
    bill = 3*units
  elif 100 < units <= 200:
    bill = 5*units
  else:
    bill = 7*units

  if bill < 500:
    usage = "Low usage"
  elif 500 < bill < 1500:
    usage = "Moderate usage"
  else:
    usage = "High usage"

  return(bill,usage)
units = int(input("Units consumed:"))
Bill(units)

```
