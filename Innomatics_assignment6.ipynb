1:Smart Parking Lot Management System

```python
def parking_system():

    capacity = int(input("Enter Parking Capacity: "))
    n = int(input("Enter number of log events: "))
    parked = 0
    peak = 0
    for i in range(n):
        event = input("Enter event (entry/exit): ").lower()

        if event == "entry":
            parked += 1
        elif event == "exit":
            parked -= 1

        if parked > peak:
            peak = parked
    print("Total Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak)
    if peak > capacity:
        print("Alert: Capacity Exceeded")
    else:
        print("Parking within capacity")
parking_system()
```

2:Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    speed = 25  # realistic city speed
    time = (distance / speed) * 60
    traffic_delay_map = {
        "low": 5,
        "medium": 10,
        "high": 20
    }
    weather_delay_map = {
        "clear": 0,
        "rainy": 15,
        "stormy": 25
    }
    traffic_delay = traffic_delay_map.get(traffic.lower(), 0)
    weather_delay = weather_delay_map.get(weather.lower(), 0)
    final_time = time + traffic_delay + weather_delay
    return round(final_time)

distance = float(input("Enter Distance (km): "))
traffic = input("Enter Traffic Level (Low / Medium / High): ")
weather = input("Enter Weather (Clear / Rainy / Stormy): ")
print(f"Estimated delivery time:{estimate_delivery_time(distance, traffic, weather)} minutes")

```

3:Movie Theatre Seat Occupancy

```python
total_seats=int(input("Total Seats:"))
def movie_theater_occupancy(x):
  occupancy_percentage = (len(x)*100)/total_seats
  print("Occupancy:",occupancy_percentage,"%")
  if len(x) == total_seats:
    print("Show Status: Housefull")
  elif total_seats > len(x) > total_seats/2 :
    print("Show Status: More than 50% filled")
  elif total_seats > len(x) < total_seats/2:
    print("Show Status: Less than 50% filled")
  else:
    print("Show Status: No Bookings yet")
x = list(map(int, input("Booked seats:").split()))
movie_theater_occupancy(x)
```

4:Cloud Server Load Classification System

```python
def server_load_classification(cpu_readings):
  avg_cpuload = (sum(cpu_readings)/len(cpu_readings))
  print(f"Average CPU Load:{avg_cpuload}%")
  if avg_cpuload < 50:
    print("Server status: Normal")
  elif 50 < avg_cpuload < 80:
    print("Server status: Warning")
  else:
    print("Server status: Critical")
cpu_readings = list(map(int, input("CPU Readings (%):").split()))
server_load_classification(cpu_readings)
```

5:Smart Classroom Resource Usage Monitor


```python
def classroom_resource_usage():
  resource_usage = {
      "Projector": 0, # Initialize with 0 hours
      "AC": 0,
      "Lights": 0
  }

  # Get input for each resource and update the dictionary
  resource_usage["Projector"] = int(input("Projector used for how many hours? "))
  resource_usage["AC"] = int(input("AC used for how many hours? "))
  resource_usage["Lights"] = int(input("Lights used for how many hours? "))

  # Find the resource with the maximum usage
  max_hours = 0
  overused_resource = ""

  for resource, hours in resource_usage.items():
    if hours > max_hours:
      max_hours = hours
      overused_resource = resource

  if overused_resource:
    print(f"Overused Resource: {overused_resource}")
  else:
    print("No resources used or an issue occurred.")

  return resource_usage

classroom_resource_usage()
```

6: Online Event Registration Capacity Controller

```python
def event_controller():

    capacity = int(input("Event Capacity: "))
    registrations = int(input("Registrations: "))
    Confirmed_registrations = 0
    waitlist_users = 0

    if registrations <= capacity:
        Confirmed_registrations = registrations
        waitlist_users = 0
    else:
        Confirmed_registrations = capacity
        waitlist_users = registrations - capacity

    print("Confirmed Registrations:", Confirmed_registrations)
    print("Waitlisted Users:",  waitlist_users)

    if  waitlist_users > 0:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")

event_controller()
```
