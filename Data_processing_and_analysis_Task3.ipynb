```python
#1. Employee Performance Bonus Eligibility

employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
Bonus = max(employees.values())
best_performers = []
for key in employees:
  if employees[key] == Bonus:
    best_performers.append(key)
print("Top Performers Eligible for Bonus:", ", ".join(best_performers), "(Score:", Bonus, ")")
```

```python
#2: Search Query Keyword Analysis

i_p = "Buy mobile phone buy phone online"
L = i_p.lower().split()
D = {}
for word in L:
    if word in D:
        D[word] += 1
    else:
        D[word] = 1
print(D)
```

```python
#3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]
Valid_readings=[]
Index=[]

for m in sensor_readings:
  if m%2 == 0:
    Valid_readings.append(m)
for n in Valid_readings:
  Index.append(sensor_readings.index(n))
pairs = list(zip(Index,Valid_readings))
print("Valid Sensor Readings (Hour, Value):\n",pairs)
```

```python
#4: Email Domain Usage Analysis

emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

Gmail=[]
yahoo=[]

for domain in emails:
  if "gmail.com" in domain:
    Gmail.append(domain)
  elif "yahoo.com" in domain:
    yahoo.append(domain)
percentage_gmail = len(Gmail)/len(emails)*100
percentage_yahoo = len(yahoo)/len(emails)*100
print("gmail.com:",percentage_gmail,"%""\n yahoo.com:",percentage_yahoo,"%")
```

```python
#5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

avg = sum(sales) / len(sales)
threshold = avg * 1.3

for day, value in enumerate(sales, start=1):
    if value >= threshold:
        print(f"Day {day}: {value}")
```

```python
#6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
repeated = []
for i in user_ids:
  if user_ids.count(i)>1 and i not in repeated:
    repeated.append(i)
for id in repeated:
    print(id, user_ids.count(id),"times")
```
